### Task 1

a)

We begin by considering the function $f(x) = e^{2 \pi i k x}$, with $k \in \mathbb{Z}$ and $x \in \mathbb{T}$, and the inner product of two functions $f, g \in \mathbb{L}^2(\mathbb{T})$ as $\langle f , g \rangle := \int_0^{1} f(x) \overline{g(x)} dx$.

For two complete functions on this form, ie. not evaluated at one specific point $x$, but rather the entire functions, we can denote the inner product as follows:

$\langle e^{2 \pi i k \cdot} , e^{2 \pi i h \cdot} \rangle = \int_0^{1} e^{2 \pi i k \cdot} \overline{e^{2 \pi i h \cdot}} d\cdot = \int_0^{1} e^{2 \pi i k \cdot} e^{-2 \pi i h \cdot} d\cdot = \int_0^{1} e^{2 \pi i (k-h) \cdot} d\cdot$

Since $k, h \in \mathbb{Z}$, we now have two special cases of this integral to consider, namely $I) \quad k = h$, and $II) \quad k \neq h$.

$I):$

For the case where $k = h$, we get that the inner product becomes:

$\langle e^{2 \pi i k \cdot} , e^{2 \pi i k \cdot} \rangle = \int_0^{1} e^{2 \pi i (k-k) \cdot} d\cdot = \int_0^{1} e^{2 \pi i (0) \cdot} d\cdot = \int_0^{1} e^{0} d\cdot = \int_0^{1} 1 d\cdot = [\cdot]_0^{1} = (1-0) = \underline{1}$

$II):$

For the second case, where $k \neq h$, as previously calculated, the inner product becomes:

$\langle e^{2 \pi i k \cdot} , e^{2 \pi i h \cdot} \rangle = \int_0^{1} e^{2 \pi i (k-h) \cdot} d\cdot$

By now substituting in the dummy variable $a \equiv 2 \pi i (k-h)$, we get:

$= \int_0^1 e^{a \cdot} d\cdot = [\frac{e^{a \cdot}}{a}]_0^1 = \frac{1}{a} (e^{a} - e^{0})$

Substitute back $2 \pi i (k-h) = a$, and we have:

$= \frac{1}{2 \pi i (k-h)} (e^{2 \pi i (k-h)} - e^{0}) = \frac{1}{2 \pi i (k-h)} (e^{2 \pi i (k-h)} - 1)$

Since both $k, h \in \mathbb{Z}$, it follows that $n \equiv k - h \in \mathbb{Z}$, and we know by Euler's formula that $e^{2 \pi i n} = 1, 
\forall n \in \mathbb{Z}$, therefore:

$= \frac{1-1}{2 \pi i (k-h)} = \frac{0}{2 \pi i (k-h)} = \underline{0}$

Combining these two results, we have that:

$\underline{\underline{ \langle e^{2 \pi i k \cdot} , e^{2 \pi i h \cdot} \rangle = 
\begin{cases}
    1 ,& \text{if } k = h\\
    0,             & k \neq h\
\end{cases}}}$



b)

Orthonormality:

We consider functions on the form $\sqrt{2}sin(2 \pi n x)$, where $m = 1, 2, ...$, $cos(2 \pi 0 x)$, and $\sqrt{2} cos(2 \pi n x)$, where $n = 1, 2,...$, and $x \in \mathbb{T}$.

In order for these three functions to form an orthonormal system, they must satisfy the following conditions:

$I:$ The inner product of different functions in the system is $0$ (orthogonality)

$II:$ The inner product of a function with itself is $1$ (normality)

These conditions can, in this case, be expressed as follows:

$I: \langle \sqrt{2}sin(2 \pi n \cdot), \sqrt{2}cos(2 \pi m \cdot) \rangle = 0, \quad \forall n \in \{1,2,...\}, m \in \{0,1,...\}$ 

$II: \langle \sqrt{2}sin(2 \pi n \cdot), \sqrt{2}sin(2 \pi m \cdot) \rangle = 
\begin{cases}
    0 , & \text{if } m \neq n\\
    1,              & m = n\
\end{cases}$

$III: \langle \sqrt{2}cos(2 \pi n \cdot), \sqrt{2}cos(2 \pi m \cdot) \rangle = 
\begin{cases}
    0 , & \text{if } m \neq n\\
    1,              & m = n \neq 0\\
    2,              & m = n = 0\
\end{cases}$

Since we now are dealing only with real valued functions, not complex ones, the defined inner product is:

$\langle f , g \rangle = \int_0^1 f(x) \overline{g(x)} dx = \int_0^1 f(x) g(x) dx$

$I):$ Orthogonality between $\sqrt{2}sin(2 \pi n \cdot)$ and $\sqrt{2}cos(2 \pi m \cdot)$

Start by computing the inner product:

$\sqrt{2}sin(2 \pi n \cdot)$ and $\sqrt{2}cos(2 \pi m \cdot) = 2 \int_0^1 sin(2 \pi n \cdot) cos(2 \pi m \cdot) d\cdot$

Now use this product-to-sum trigonomic identity: $sin(A)cos(B) = \frac{1}{2}[sin(A+B)+sin(A-B)]$:

$\Rightarrow 2 \int_0^1 \frac{1}{2} [sin(2 \pi n \cdot + 2 \pi m \cdot) + sin(2 \pi n \cdot - 2 \pi m \cdot)] d\cdot$

$= \int_0^1 sin(2 \pi (n + m) \cdot)d\cdot + \int_0^1 sin(2 \pi (n - m) \cdot) d\cdot$

Since both $n \in \{1,2,... \} m \in \{0,1,... \}$, it follows that both $n+m$ and $n-m$ are integers, and it is known that for any integer $k$ the integral $\int_0^1 sin(2 \pi k x) dx = 0$, because the integral over an integer number of periods cancels out. Therefore:

$\underline{\underline{\langle \sqrt{2}sin(2 \pi n \cdot), \sqrt{2}cos(2 \pi m \cdot) \rangle = \int_0^1 sin(2 \pi (n + m) \cdot)d\cdot + \int_0^1 sin(2 \pi (n - m) \cdot) d\cdot = 0+0 = 0}}$

ie. there is orthogonality between different functions on the forms $\sqrt{2}sin(2 \pi n \cdot)$ and $\sqrt{2}cos(2 \pi m \cdot)$

$II):$ 

Compute the inner product:

$\langle \sqrt{2}sin(2 \pi n \cdot), \sqrt{2}sin(2 \pi m \cdot) \rangle = \int_0^1 \sqrt{2}sin(2 \pi n \cdot) \sqrt{2}sin(2 \pi m \cdot) d\cdot = 2 \int_0^1 sin(2 \pi n \cdot) sin(2 \pi m \cdot) d\cdot$

Use the following product-to-sum trigonomic identity: $sin(a)sin(B) = \frac{1}{2}[cos(A-B) - cos(A+B)]$

$\Rightarrow 2 \int_0^1 sin(2 \pi n \cdot) sin(2 \pi m \cdot) d\cdot = \int_0^1 cos(2 \pi n \cdot - 2 \pi m \cdot) d\cdot - \int_0^1 cos(2 \pi n \cdot + 2 \pi m \cdot) d\cdot = \int_0^1 cos(2 \pi (n - m) \cdot) d\cdot - \int_0^1 cos(2 \pi (n + m) \cdot) d\cdot$

Again, $n-m$ and $n+m$ are both integers, and the integral of cosine for a non-zero integer of times is $0$. Thus:

$n \neq m$:

$\int_0^1 cos(2 \pi (n - m) \cdot) d\cdot - \int_0^1 cos(2 \pi (n + m) \cdot) d\cdot = 0-0 = 0$

ie. there is orthogonality between different functions on the form $\sqrt{2}sin(2 \pi n \cdot)$

$n = m$:

$\int_0^1 cos(2 \pi (n - m) \cdot) d\cdot - \int_0^1 cos(2 \pi (n + m) \cdot) d\cdot = \int_0^1 cos(2 \pi (0) \cdot) d\cdot - 0 = \int_0^1 1 d\cdot = 1$

ie. the functions on the form $\sqrt{2}sin(2 \pi n \cdot)$ are normalized

$\underline{\underline{\Rightarrow \langle \sqrt{2}sin(2 \pi n \cdot), \sqrt{2}sin(2 \pi m \cdot) \rangle = 
\begin{cases}
    0 , & \text{if } m \neq n\\
    1,              & m = n\
\end{cases}}}$

$III):$ 

Compute the inner product:

$\langle \sqrt{2}cos(2 \pi n \cdot), \sqrt{2}cos(2 \pi m \cdot) \rangle = \int_0^1 \sqrt{2}cos(2 \pi n \cdot) \sqrt{2}cos(2 \pi m \cdot) d\cdot = 2 \int_0^1 cos(2 \pi n \cdot) cos(2 \pi m \cdot) d\cdot$

This time use the trigopnomic identity as follows: $cos(A)cos(B) = \frac{1}{2}[cos(A-B) + cos(A+B)]$

$\Rightarrow 2 \int_0^1 cos(2 \pi n \cdot) cos(2 \pi m \cdot) d\cdot = \int_0^1 cos(2 \pi n \cdot - 2 \pi m \cdot) + cos(2 \pi n \cdot + 2 \pi m \cdot) d\cdot = \int_0^1 cos(2 \pi (n - m) \cdot) d\cdot + \int_0^1 cos(2 \pi (n + m) \cdot) d\cdot$

$n \neq m$:

For $n \neq m$, both terms cancel out as previously explained, and we are left with $\int_0^1 cos(2 \pi (n - m) \cdot) d\cdot + \int_0^1 cos(2 \pi (n + m) \cdot) d\cdot = 0 + 0 = 0$

$n = m \neq 0$:

In this case, we have that the second term $\int_0^1 cos(2 \pi (n + m) \cdot) d\cdot = 0$ as previously explained, whereas the first term $\int_0^1 cos(2 \pi (n - m) \cdot) d\cdot = \int_0^1 cos(0) d\cdot = \int_0^1 1 d\cdot = [\cdot]_0^1 = (1-0) = 1$

$n = m = 0$:

$\int_0^1 cos(2 \pi (n - m) \cdot) d\cdot + \int_0^1 cos(2 \pi (n + m) \cdot) d\cdot = \int_0^1 cos(0) d\cdot + \int_0^1 cos(0) d\cdot = 2 \times [\cdot]_0^1 = 2$

$\underline{\underline{\Rightarrow \langle \sqrt{2}cos(2 \pi n \cdot), \sqrt{2}cos(2 \pi m \cdot) \rangle = 
\begin{cases}
    0 , & \text{if } m \neq n\\
    1,              & m = n \neq 0\\
    2,              & m = n = 0\
\end{cases}}}$

c)
Orthonormal basis

We define two spaces $T_n$ and $S_n$:

$$T_n := span(e^{-2 \pi i n \cdot},...,e^{2 \pi i n \cdot}) = \{ f | f(x) = \sum_{k=-n}^n c_k e^{2 \pi i k x}, \quad where \quad c_{-n},...,c_n \in \mathbb{C}\}$$

and 

$$S_n := span(cos(0 \cdot),cos(2 \pi \cdot),...,cos(2 \pi n \cdot),sin(2 \pi \cdot),sin(2 \pi 2 \cdot),...,sin(2 \pi n \cdot)) = \{ f | f(x) = \frac{a_0}{2} + \sum_{k=1}^n ( a_k cos(2 \pi k x) + b_k sin(2 \pi k x)), \quad where \quad a_{0},a_{1},...,a_n,b_{1},...,b_n \in \mathbb{R} \}$$

We also have the restriction that $c_k = \overline{c_{-k}}$.

Euler's identity states that:

$$e^{2 \pi i k \cdot} = cos(2 \pi k \cdot) + i sin(2 \pi k \cdot)$$
$$e^{-2 \pi i k \cdot} = cos(2 \pi k \cdot) - i sin(2 \pi k \cdot)$$
$$\forall f(x) \in T_n$$

We decompose Euler's identity into terms with positive and negative values of $k$, and $k=0$ by itself:

$k=0: \quad c_0e^{2 \pi i (0) \cdot} = c_0$

$k > 0:$

For $k > 0$ we use the following decomposition to represent it in terms of conjugate pairs:

$c_k e^{2 \pi i k \cdot} + c_{-k} e^{-2 \pi i k \cdot} = c_k(cos(2 \pi k \cdot) + i sin(2 \pi k \cdot)) + c_{-k}(cos(2 \pi k \cdot) - i sin(2 \pi k \cdot))$

Now exploit the restriction of $c_k = \overline{c_{-k}} \Rightarrow c_{-k} = \overline{c_k}$, and group real and imaginary terms together:


$= c_k(cos(2 \pi k \cdot) + i sin(2 \pi k \cdot)) + \overline{c_k}(cos(2 \pi k \cdot) - i sin(2 \pi k \cdot)) = (c_k + \overline{c_k})cos(2 \pi k \cdot) + i(c_k - \overline{c_k}) sin(2 \pi k \cdot)$

As previously mentioned we have that $c_k = \overline{c_{-k}}$, which means that the term $(c_k + \overline{c_{-k}}) = a_k$, where $a_k = 2Re(c_k) \in \mathbb{R}$, as the imaginary parts cancel each other out.

As for $(c_k - \overline{c_k}) = 2Im(c_k)$, as the real parts cancel out, and as such $i(c_k - \overline{c_k}) = b_k \in \mathbb{R}$

For $c_0 = \overline{c_0}$ to hold, we must have that $c_0 \in \mathbb{R}$

Combining all of this, we have that:

$\underline{\underline{f(x) = \frac{a_0}{2} + \sum_{k=1}^n ( a_k cos(2 \pi k x) + b_k sin(2 \pi k x))}}, \quad$ where $a_0 = 2c_0 \in \mathbb{R}, a_k = 2Re(c_k) \in \mathbb{R}, b_k = i2Im(c_k) \in \mathbb{R}$

We observe that this is a function written on the form of $S_n$ and therefor deduct that $T_n = S_n$, they are merely expressed in two different ways, as complex exponentials and real trigonometric functions respectively.

$\underline{Dim(T_n)}$:

$T_n$ has $n$ dimensions for strictly positive values of $k$, and $n$ dimensions for strictly negative values of $k$, as well as one dimension for $k=0$:

In total: $\underline{\underline{Dim(T_n) = 2n+1}}$

This is also the same as the dimensions of $S_n$ which has $n$ dimensions corresponding to the sine-functions, and $n+1$ for cosine.

d)

We define the standard inner product as:

$\langle f, g \rangle = \int_0^1 f(x) \overline{g(x)} dx = \int_0^1 f(x) g(x) dx$, because we are dealing only with real functions

In order to calculate the Fourier-coefficients $a_k$ and $b_k$, by definition we use the inner products with $cos(2 \pi k \cdot)$ and $sin(2 \pi k \cdot)$ respectively:

$\underline{a_k (k \geq 1)}$:

$\langle f, cos(2 \pi k \cdot) \rangle = \int_0^1 (\frac{a_0}{2} + \sum_{j=1}^n ( a_j cos(2 \pi j \cdot) + b_j sin(2 \pi j \cdot))) cos(2 \pi k \cdot) d\cdot = \int_0^1 \frac{a_0}{2}cos(2 \pi k \cdot) + \sum_{j=1}^n  a_j cos(2 \pi j \cdot)cos(2 \pi k \cdot) + \sum_{j=1}^n b_j sin(2 \pi j \cdot)cos(2 \pi k \cdot)  d\cdot$

We now restate the orthonormality statements from b as follows:

$\langle \sqrt{2}cos(2 \pi n \cdot), \sqrt{2}cos(2 \pi m \cdot) \rangle = 
\begin{cases}
    0 , & \text{if } m \neq n\\
    1,              & m = n \neq 0\\
    2,              & m = n = 0\
\end{cases} \quad \Rightarrow \quad \langle cos(2 \pi n \cdot), cos(2 \pi m \cdot) \rangle = 
\begin{cases}
    0 , & \text{if } m \neq n\\
    \frac{1}{2},              & m = n \neq 0\\
    1,              & m = n = 0\
\end{cases}$

From this, we can see that the first and last terms vanish due to orthogonality, and we are left with:

$\langle f, cos(2 \pi k \cdot) \rangle = \int_0^1 \sum_{j=1}^n a_j cos(2 \pi j \cdot)cos(2 \pi k \cdot) d\cdot = \sum_{j=1}^n \langle a_j cos(2 \pi j \cdot), cos(2 \pi k \cdot) \rangle $, which, also from orthogonality, the inner product is $0$ for all $j \neq k$, and is $\frac{1}{2}$ for all $j = k \neq 0$:


$\langle f, cos(2 \pi k \cdot) \rangle = \frac{1}{2} a_k \quad \Rightarrow \quad \underline{\underline{a_k = 2 \langle f, cos(2 \pi k \cdot) \rangle}}$

$\underline{a_k (k = 0)}$:

For $a_0$, much of the calculations are the same, but we now have the following case:

$\langle f, cos(2 \pi (0) \cdot) \rangle = \langle f, 1 \rangle = \int_0^1 (\frac{a_0}{2} + \sum_{j=1}^n ( a_j cos(2 \pi j \cdot) + b_j sin(2 \pi j \cdot)) 1 d\cdot = \langle \frac{a_0}{2}, 1 \rangle + \sum_{j=1}^n \langle  a_j cos(2 \pi j \cdot), 1 \rangle +  \sum_{j=1}^n \langle  b_j sin(2 \pi j \cdot), 1 \rangle$ 

This time it is the last two terms that disappear due to orthogonality, and we are left with:

$\langle f, cos(2 \pi (0) \cdot) \rangle = \langle \frac{a_0}{2}, 1 \rangle = \frac{a_0}{2} \quad \Rightarrow \quad a_0 = 2 \langle f, cos(2 \pi (0) \cdot)$, which follows the same pattern as all other $k$, namely:

$\underline{\underline{a_k = 2 \langle f, cos(2 \pi k \cdot) \rangle}}$

$\underline{b_k}$:

In order to calculate the $b_k$-coefficients, the method is the same, but this time we consider the inner product of $f$ and $sin(2 \pi k \cdot)$:

$\langle f, sin(2 \pi k \cdot) \rangle = \int_0^1 (\frac{a_0}{2} + \sum_{j=1}^n ( a_j cos(2 \pi j \cdot) + b_j sin(2 \pi j \cdot))) sin(2 \pi k \cdot) d\cdot = \int_0^1 \frac{a_0}{2}sin(2 \pi k \cdot) + \sum_{j=1}^n  a_j cos(2 \pi j \cdot)sin(2 \pi k \cdot) + \sum_{j=1}^n b_j sin(2 \pi j \cdot)sin(2 \pi k \cdot)  d\cdot$

This time, the first two terms are $0$ due to orthogonality, whereas the last term is:

$\langle f, sin(2 \pi k \cdot) \rangle = \int_0^1 \sum_{j=1}^n b_j sin(2 \pi j \cdot)sin(2 \pi k \cdot) d\cdot = \sum_{j=1}^n \langle b_j sin(2 \pi j \cdot), sin(2 \pi k \cdot) \rangle $, which, also from orthogonality, this time between the different sine-functions instead of cosine, the inner product is $0$ for all $j \neq k$, and is $\frac{1}{2}$ for all $j = k$.

$\langle f, sin(2 \pi k \cdot) \rangle = \frac{1}{2} b_k \quad \Rightarrow \quad \underline{\underline{b_k = 2 \langle f, sin(2 \pi k \cdot) \rangle}}$

e)

The Fourier-coefficients $c_k$ for a function $f$ on the interval $[0,1]$ is known to be given by:

$c_k = \langle f, e^{2 \pi i k x} \rangle = \int_0^1 f(x) e^{-2 \pi i k x} dx$

The composite trapezoidal rule for an $N$ number of equidistant points, is:

$\int_a^b f(x) dx \approx  \frac{h}{2} \sum_{j=0}^{N-1} f(x_j) + f(x_{j+1}) = \frac{h}{2} (f(x_0) + 2f(x_1) + ... + 2f(x_{N-1}) + f(x_N))$, where $x_j = \frac{j}{N}$, and $j=0,...,N-1$ and $h=\frac{(b-a)}{N}$

Now we assume that $f(x)$ is periodic, and therefore that $f(x_N) = f(x_0)$, plug in $h=\frac{(b-a)}{N} = \frac{(1-0)}{N}$ and simplify:

$\Rightarrow \frac{h}{2} (2f(x_0) + 2f(x_1) + ... + 2f(x_{N-1})) = \frac{1}{N} (f(x_0) + f(x_1) + ... + f(x_{N-1})) = \frac{1}{N} \sum_{j=0}^{N-1} f(x_j)$

By applying the trapezoidal approximation to the function for the Fourier-coefficients $c_k$, we get:

$c_k(f) = \int_0^1 f(x) e^{-2 \pi i k x} \approx \hat{f}_k = \frac{1}{N} \sum_{j=0}^{N-1} f(x_j) e^{-2 \pi i k x_j}$

Now let $x_j = \frac{j}{N}$, and $j=0,...,N-1$, and $f_j := f(x_j) = f(\frac{j}{N})$:

$\Rightarrow \underline{\underline{\hat{f_k} = \frac{1}{N} \sum_{j=0}^{N-1} f_j e^{\frac{-2 \pi i k j}{N}}}}$

Periodicity of *$\hat{f_k}$*:

For *$\bold{\hat{f_k}}$* to be periodic with period $N$, the condition $\hat{f_k} = \hat{f}_{k+N}$ must hold.

$\hat{f}_{k+N} = \frac{1}{N} \sum_{j=0}^{N-1} f_j e^{\frac{-2 \pi i (k+N) j}{N}}$

$= \frac{1}{N} \sum_{j=0}^{N-1} f_j e^{\frac{-2 \pi i k j}{N}} e^{\frac{-2 \pi i N j}{N}} = \frac{1}{N} \sum_{j=0}^{N-1} f_j e^{\frac{-2 \pi i k j}{N}} e^{-2 \pi i j}$

Since $e^{-2 \pi i j} = 1 \forall j \in \mathbb{Z}$, this becomes:

$\underline{\underline{\hat{f}_{k+N} = \frac{1}{N} \sum_{j=0}^{N-1} f_j e^{\frac{-2 \pi i k j}{N}} = \hat{f}_k}}$

The fact that the approximation $\hat{f}_k$ is periodic, means that the Fourier coefficients $c_k$ also repeat every $N$ values.

f)

We let $N \in \mathbb{N}$ and $k \in \mathbb{Z}$, and have the sum:

$\frac{1}{N} \sum_{j=0}^{N-1} e^{-2 \pi i j \frac{k}{N}}$

Now consider two special cases of this sum:

$I) \quad k \mod N = 0 \quad$ and $\quad II) \quad k \mod N \neq 0$

$I)$:

$k \mod N = 0 \Leftrightarrow \frac{k}{N} = m \Rightarrow k=mN \quad$, where $m \in \mathbb{Z}$

$\Rightarrow \frac{1}{N} \sum_{j=0}^{N-1} e^{-2 \pi i j \frac{k}{N}} = \frac{1}{N} \sum_{j=0}^{N-1} e^{-2 \pi i j \frac{mN}{N}} = \frac{1}{N} \sum_{j=0}^{N-1} e^{-2 \pi i j m} = \frac{1}{N} \sum_{j=0}^{N-1} 1 = \frac{1}{N}N = \underline{1}$

$II)$:

$\frac{1}{N} \sum_{j=0}^{N-1} e^{-2 \pi i j \frac{k}{N}} = \frac{1}{N} \sum_{j=1}^{N} e^{-2 \pi i (j-1) \frac{k}{N}} = \frac{1}{N} \sum_{j=1}^{N} (e^{-2 \pi i \frac{k}{N}})^{(j-1)}$

Recognise this sum as a geometric series:

$\frac{1}{N} \sum_{j=1}^{N} ar^{(j-1)} = a \frac{(1-r^N)}{1-r} \quad$, with $a=1$, and $r = e^{-2 \pi i \frac{k}{N}}$

$\Rightarrow \frac{1}{N} \sum_{j=0}^{N-1} e^{-2 \pi i j \frac{k}{N}} = 1 \frac{(1-(e^{-2 \pi i \frac{k}{N}})^N)}{(1-e^{-2 \pi i \frac{k}{N}})} = \frac{1-e^{-2 \pi i k}}{1-e^{-2 \pi i \frac{k}{N}}} = \frac{1-1}{1-e^{-2 \pi i \frac{k}{N}}} = \underline{0}$

Combining these two results, we have that:

$\underline{\underline{\frac{1}{N} \sum_{j=0}^{N-1} e^{-2 \pi i j \frac{k}{N}} = \begin{cases}
    1 , & \text{if } k \mod N \equiv 0\\
    0,              & \text{else}\
\end{cases}}}$

g)

$F_N = (e^{-2 \pi i \frac{kl}{N}})_{k,l}^{N-1} \quad \Rightarrow \quad \overline{F_N} = (e^{2 \pi i \frac{lk}{N}})_{k,l}^{N-1}$

Calculate the matrix product of the two:

$F_N \overline{F_N} = (e^{-2 \pi i \frac{kl}{N}})_{k,l}^{N-1} (e^{2 \pi i \frac{lk'}{N}})_{k',l}^{N-1}$

Now consider one single element of the product matrix:

$(F_N \overline{F_N})_{k,k'} = \sum_{l=0}^{N-1} e^{-2 \pi i \frac{kl}{N}} e^{2 \pi i \frac{lk'}{N}} = \sum_{l=0}^{N-1} e^{2 \pi i l \frac{(k'-k)}{N}}$

As shown in the last task, this sum is:

$\sum_{l=0}^{N-1} e^{2 \pi i l \frac{(k'-k)}{N}} = \begin{cases}
    N , & \text{if } (k'-k) \mod N \equiv 0  \quad \Rightarrow \quad  k' = k\\
    0,              & \text{else}\
\end{cases}$

$(k'-k) \mod N \equiv 0  \quad \Rightarrow \quad  k' = k$ being true, since neither $k'$ or $k$ can exceed $N$

$\Rightarrow (F_N \overline{F_N})_{k,k'} = \begin{cases}
    N , &  \text{if } k' = k\\
    0,              & \text{else}\
\end{cases}$

$\Rightarrow F_N \overline{F_N} = N \mathbf{I} \quad \Rightarrow \quad F_N \frac{1}{N} \overline{F_N} = \mathbf{I} \quad$, where $\mathbf{I}$ is the identity matrix of size $N$$

By definition, the inverse is:

$F_N F_N^{-1} = \mathbf{I}$

Combining these two statements, we can clearly see that:

$\underline{\underline{F_N^{-1} = \frac{1}{N} \overline{F_N}}}$

By using the scaled version $\frac{1}{\sqrt{N}} F_N$, we have that:

$(\frac{1}{\sqrt{N}} F_N) (\frac{1}{\sqrt{N}} \overline{F_N}) = \mathbf{I}$

This leads to the rather fortunate case where it is possible to calculate the inverse of the original $F_N$ matrix simply by finding the complex conjugate matrix, ie. transposing the matrix and taking the complex conjugate of each of the entries, which is much easier than calculating said inverse by traditional methods. A matrix like this, where $A\overline{A} = \mathbf{I} \Rightarrow \overline{A} = A^{-1}$ is called a unitary matrix.

We introduce the notion of the curculant matrix of a vector $a$ as follows:

$circ(a) = (a_{k-l \mod N})_{k,l=0}^{N-1}$

and also remember that:

$(F_N)_{k,l} = e^{-2 \pi i \frac{kl}{N}} \quad \Rightarrow \quad (\overline{F_N})_{k,l} = e^{2 \pi i \frac{lk}{N}}$

and

$F_N^{-1} = \frac{1}{N} \overline{F_N}$

The Fourier transformation $\hat{a}$ of a vector $a$ is a vector consisting of it's Fourier-coefficients, and defined in the following way, with $\hat{a}_k$ being it's $k$'th entry:

$\hat{a} = F_N a \quad \Rightarrow \quad \hat{a}_k = \sum_{n=0}^{N-1} a_n e^{-2 \pi i \frac{nk}{N}}$

We now consider the matrix product of:

$F_N^{-1} diag(\hat{a}) F_N$

and in particaular the $k,l$'th entry of it. We start by decomposing this into:

$(F_N^{-1} diag(\hat{a}) F_N)_{k,l} = \sum_{m=0}^{N-1}(F_N^{-1})_{k,m} (diag(\hat{a}) F_N)_{m,l}$

By the definitions, we have that the part:

$(diag(\hat{a}) F_N)_{m,l} = \hat{a}_m e^{-2 \pi i \frac{ml}{N}} \quad$ and $\quad (F_N^{-1})_{k,m} = \frac{1}{N} e^{2 \pi i \frac{km}{N}}$

$\Rightarrow (F_N^{-1} diag(\hat{a}) F_N)_{k,l} = \sum_{m=0}^{N-1}\frac{1}{N} e^{2 \pi i \frac{km}{N}} \hat{a}_m e^{-2 \pi i \frac{ml}{N}} = \frac{1}{N} \sum_{m=0}^{N-1} \hat{a}_m e^{2 \pi i m \frac{k-l}{N}}$

By substituting in the expression for $\hat{a}_m$, we get:

$\Rightarrow (F_N^{-1} diag(\hat{a}) F_N)_{k,l} = \frac{1}{N} \sum_{m=0}^{N-1} \sum_{n=0}^{N-1} a_n e^{-2 \pi i \frac{nm}{N}} e^{2 \pi i m \frac{k-l}{N}} = \frac{1}{N} \sum_{n=0}^{N-1} a_n \sum_{m=0}^{N-1} e^{2 \pi i m \frac{k-l-n}{N}}$  

As previously shown in task f, the sum $\sum_{m=0}^{N-1} e^{2 \pi i m \frac{k-l-n}{N}} = \begin{cases}
    N , & \text{if } k - l - n \mod N \equiv 0 \Rightarrow n \equiv k - l \mod N\\
    0,              & \text{else}\
\end{cases}$

and therefore, the only non-zero contributing term to this sum is where $n \equiv k - l \mod N$, and the first sum is also reduced to:

$\underline{\underline{(F_N^{-1} diag(\hat{a}) F_N)_{k,l} = \frac{1}{N} a_{k-l \mod N} N = a_{k-l \mod N}}}$

See here that all the entries of $F_N^{-1} diag(\hat{a}) F_N$ matches the definition of the circular matrix, and therefore conclude that the Fourier matrix diagonalises the circular matrix.

The periodic convolution between two functions $f$ and $g$ is defined as follows:

$(f \star g)(x) = \int_0^1 f(y) g(x-y)dy$

For two vectors $a = [a_{0}, a_1,...,a_{N-1}]$ and $b = [b_{0}, b_1,...,b_{N-1}]$ we can do the same and get the result that each entry $j$ of the convolution is:

$(c)_j = (a \star b) = \sum_{k=0}^{N-1} a_k b_{j-k \mod N}$

We now introduce the shifted version of $b$, $b' = [b_{N-1}, b_0,...,b_{N-2}]$, and also the shifted verion of $c$, corresponding to the convolution using $b'$:

$(c')_j = (a \star b') = \sum_{k=0}^{N-1} a_k b'_{j-k \mod N} = \sum_{k=0}^{N-1} a_k b_{j-k-1 \mod N} = (a \star b)_{j+1 \mod N}$

This means that the convolution $c' = (a \star b')$ is a cyclically shifted version of the original $c$, with:

$\underline{\underline{(c')_j = (c)_{j+1 \mod N}}}$

We want to convolve the sampled signal of the function $f_5(x)$ with the de-la-Valle-Poussin mean $V_{n,m}$ with values $n=48, m=1$, ie. the modified Dirichlet kernel $D_{48}$.

$f_5(x) = sin(32 \pi x) + cos(128 \pi x)$



```python
N = 512
n = 48
m = 1
x = np.linspace(0, 1, N, endpoint=False) 

def f5(x):
    return np.sin(32 * np.pi * x) + np.cos(128 * np.pi * x)

def Vnm(x, n, m):
    Vnm_values = (1 / (2 * n)) * (
        1 + 2 * np.sum([np.cos(2 * np.pi * k * x) for k in range(1, n - m + 1)], axis=0) +
        2 * np.sum([(m + n - k) / (2 * m) * np.cos(2 * np.pi * k * x) for k in range(n - m + 1, n + m)], axis=0)
    )
    return Vnm_values


# Compute the FFT's
f5_fft = np.fft.fft(f5(x))
Vnm_fft = np.fft.fft(Vnm(x, n, m))

conv_fft = f5_fft * Vnm_fft # Element-wise multiplication

filtered_sig = np.fft.ifft(conv_fft) # Take the inverse FFT of convoluted signal

filtered_signal = np.real(filtered_sig) # Consider only the real part


plt.figure(figsize=(14, 7))

plt.subplot(2, 1, 1)
plt.plot(x, f5(x), label="Original signal f_5(x)")
plt.title("$f_5(x) = sin(32 \pi x) + cos(128 \pi x)$")
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x, filtered_signal, label="Filtered Signal", color='orange')
plt.title("Signal filtered with $V_{48,1}$ kernel")
plt.legend()

plt.tight_layout()
plt.show()
```

The function $f_5$ consists of two frequency components, $sin(32 \pi x)$, which represents the lower frequency, and $cos(128 \pi x)$, the higher frequency. When we convolve $f_5$ with the de la Valle-Poussin kernel, the signal is smoothed. This happens because the de la Valle-Poussin kernel acts as a low-pass filter, meaning it suppresses the higher frequency component $cos(128 \pi x)$, while allowing the lower frequency $sin(32 \pi x)$ to pass through more easily. As a result, the low-frequency component contributes more to the new, filtered signal. 

The parameter $m$ controls how much the higher frequency is suppressed. If $m$ is increased, the filter becomes less aggressive, allowing more of the higher frequency component to pass through. Consequently, the filtered signal retains more of its original high-frequency content, and the overall smoothing effect is reduced.
```python
import pandas as pd
from scipy.signal import convolve

# Loading file and get signal data
signal_data = pd.read_csv('data/project1-signal-data.csv')
signal = signal_data.iloc[:, 1].values

# Set params and kernels
N = len(signal)
t = np.linspace(0, 1, N)

Vnm_fft = np.fft.fft(Vnm(t, 92, 10)) # V_(92,10) kernel
h = np.array([-1, 2, -1] + [0] * (N - 3)) # h kernel

# Convolution using 'same' to maintain the signal length
conv_v = convolve(signal, Vnm_fft, mode='same')
conv_h = convolve(signal, h, mode='same')


plt.figure(figsize=(12, 8))


plt.subplot(3, 1, 1)
plt.plot(t, signal, label="Original Signal", color='blue')
plt.title("Original Signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)


plt.subplot(3, 1, 2)
plt.plot(t, conv_v, label="Convolution with V_{92,10}", color='green')
plt.title("Convolution with $V_{92,10}$")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)


plt.subplot(3, 1, 3)
plt.plot(t, conv_h, label="Convolution with h", color='red')
plt.title("Convolution with h")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid(True)


plt.tight_layout()
plt.show()
```

The convolution with $V_{92,10}$ is acting as a low-pass filter, which means that the higher frequencies are removed. From the plot we can therefore see the effects of the lower frequency parts of the original signal.

The convolution with the kernel h seems to mark points of great change in fluctuations in the original signal.
